## Import Library

```python
import pandas as pd
import numpy as np
import joblib
```

## Load model

```python
with open('pipeline_knn.pkl', 'rb') as file_1:
  full_imbhandling = joblib.load(file_1)
```

## Generate random data

```python
random = np.random.default_rng(seed=2895)
id = random.integers(67,72940,size=20)
gender = random.choice(['Female','Male','Other'],size=20)
age	= np.random.choice(np.arange(0.08,82.0,1.00),size=20)
hypertension = random.choice([0,1],size=20)
heart_disease = random.choice([0,1],size=20)
ever_married = random.choice(['Yes','No'],size=20)
work_type = random.choice(['Private','Self-employed','Govt_job','children','Never_worked'],size=20)
Residence_type = random.choice(['Urban','Rural'],size=20)
avg_glucose_level = np.random.choice(np.arange(55.12,271.74,0.1),size=20)
bmi = np.random.choice(np.arange(10.3,97.6,0.1),size=20)
smoking_status = random.choice(['formerly smoked','never smoked','smokes','Unknown'],size=20)

data_inf = pd.DataFrame({'id' : id,
                        'gender' : gender,
                        'age' : age,
                        'hypertension' : hypertension,
                        'heart_disease' : heart_disease,
                        'ever_married' : ever_married,
                        'work_type' : work_type,
                        'Residence_type' : Residence_type,
                        'avg_glucose_level' : avg_glucose_level,
                        'bmi' : bmi,
                        'smoking_status' : smoking_status                        
                        })
data_inf
```

## Predict

```python
full_imbhandling
```

```python
y_pred_inf= full_imbhandling.predict(data_inf)
y_pred_inf
```
